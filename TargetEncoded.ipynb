```python
import seaborn as sns
```

```python
data_sudhir=sns.load_dataset('titanic')
```

```python
data_sudhir.head()
```

```python
mean_fare=data_sudhir.groupby('sex')['fare'].mean()
```

```python
mean_fare
```

```python
mean_fare1=data_sudhir.groupby('embarked')['fare'].mean()
```

```python
mean_fare1
```

```python
mean_fare2=data_sudhir.groupby('class')['fare'].mean()

mean_fare3=data_sudhir.groupby('who')['fare'].mean()

mean_fare4=data_sudhir.groupby('embark_town')['fare'].mean()

mean_fare2,mean_fare4,mean_fare3
```

```python
data_sudhir['fare_encoded']=data_sudhir['sex'].map(mean_fare)
data_sudhir['fare_encoded1']=data_sudhir['embarked'].map(mean_fare1)
data_sudhir['fare_encoded2']=data_sudhir['class'].map(mean_fare2)
data_sudhir['fare_encoded3']=data_sudhir['who'].map(mean_fare3)
data_sudhir['fare_encoded4']=data_sudhir['embark_town'].map(mean_fare4)
```
